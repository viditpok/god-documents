### 5. Feature selection and change-point detection

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(6740)

n1, n2 = 100, 50
X_pre = np.random.normal(0, 1, n1)
X_post = np.random.normal(0, np.sqrt(1.25), n2)
X = np.concatenate([X_pre, X_post])

log_likelihood_ratio = (X**2 / 2) * (1 - 1/1.25) + 0.5 * np.log(1/1.25)

S = np.zeros(len(X))
for i in range(1, len(X)):
    S[i] = max(0, S[i-1] + log_likelihood_ratio[i])
```

```python
plt.figure(figsize=(10, 5))
plt.plot(S, label="CUSUM Statistic", color="blue")
plt.axvline(x=100, linestyle="--", color="red", label="True Change Point")
plt.xlabel("Sample Index")
plt.ylabel("CUSUM Statistic")
plt.title("CUSUM Statistic for Change Detection")
plt.legend()
plt.grid(True)
plt.show()
```
